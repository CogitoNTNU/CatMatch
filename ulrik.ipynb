```python
df.groupby("breed").agg(age_mode=("age", "count"))
```

```python
breeds = df["breed"].unique().tolist()
rows = []
for breed in breeds:
    row = df[df["breed"] == breed][["breed", "age"]].mode()
    rows.append(row)

df_new = pd.concat(rows)
df_new
```
